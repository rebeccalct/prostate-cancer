```python
TRAIN_FILE = 'Dataset_cedric/Train.csv'
SYNTHETIC_FILE = 'Dataset_cedric/SYNTHETIC DATASETS/CTGAN-2.csv'
```

## 1. Load the dataset

```python
real_data = pd.read_csv(TRAIN_FILE)
real_data
```

```python
real_data.dtypes
```

```python
discrete_columns = ["cT(MRI)","Multifocaal", "pN", "1c (0) - Perifeer (1) -  Beide (2) - Transitie (3)"]
discrete_columns
```

```python
ctgan = CTGAN()
ctgan.fit(real_data, discrete_columns)
```

```python
synthetic_data = ctgan.sample(len(real_data))
synthetic_data = synthetic_data.round(2)
synthetic_data
```
